```python
import pandas as pd

# 读取txt文件（假设列之间用制表符\t分隔）
df = pd.read_csv("taskB-loss.txt", sep="\t")
steps = df["Step"].values
loss = df["Training Loss"].values


import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator  # 用于优化横轴刻度

plt.figure(figsize=(10, 5))
plt.plot(steps, loss, color='blue', linewidth=1, label="Training Loss")

# 设置坐标轴
plt.xlabel("Steps", fontsize=12)
plt.ylabel("Loss", fontsize=12)
plt.title("Training Loss Curve", fontsize=14)

# 强制横轴显示为整数刻度
ax = plt.gca()
ax.xaxis.set_major_locator(MaxNLocator(integer=True))  # [3,9](@ref)

# 添加辅助元素
plt.grid(alpha=0.3, linestyle="--")
plt.legend()
plt.tight_layout()  # 自动调整布局

# 保存图片（分辨率300dpi）
plt.savefig("loss_curve.png", dpi=300)
plt.show()
```
